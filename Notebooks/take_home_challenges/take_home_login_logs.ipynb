```python
import pandas as pd
```

```python
int_chlng = pd.read_json('/Users/pallavibothra/Downloads/ultimate_challenge/logins.json') 
```

```python
login_freq = int_chlng.groupby(pd.Grouper(key="login_time", freq="15T", closed = 'left')).size().to_frame('total_logins')
```

```python
login_freq = login_freq.reset_index()
```

```python
#login_freq['login_time(hour)'] = login_freq['login_time'].dt.hour
login_freq['login_time(time)'] = login_freq['login_time'].dt.time
#login_freq['login_time(date)'] = login_freq['login_time'].dt.date
```

```python
login_freq.groupby('login_time(time)')['total_logins'].sum().sort_index().plot()
```

From the daily cycle plot, it can be seen the number of logins are very high around 1am, 12 pm and 10 pm. And the lowest logins are found at around 6-8 am. There are moderate number of logins around 3-5 pm.
